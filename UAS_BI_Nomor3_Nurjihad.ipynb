```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_data = pd.read_csv('F:/tugas/UAS_Business_Intellegent/caesarian.csv',delimiter=',')
```

```python
df_data.head()
```

```python
import math

dis = []
for i in range(80):
    dis.append(math.sqrt((float(df_data.iloc[i]['Usia'])-25)**2+
                         (float(df_data.iloc[i]['Kelahiran_ke-'])- 1)**2+
                         (float(df_data.iloc[i]['Waktu_Kelahiran'])-0)**2+
                         (float(df_data.iloc[i]['Tekanan_darah'])-1)**2))
```

```python
df_data['dis'] = dis
df_data
```

```python
df_data.sort_values('dis')
```

```python
y = df_data.sort_values('dis').head(5)
y
```

```python
z = y["Caesarian"]
z
```

```python
np.mean(z)
```
